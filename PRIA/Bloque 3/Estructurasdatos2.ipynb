Realizar un programa en Python que pida al usuario la dimensión y los elementos de dos listas de números enteros (pueden tener distinta dimensión):

- Una vez leídos los datos y almacenados en las listas, se generará una nueva lista que será la intersección de ambas, es decir, contendrá aquellos elementos que se encuentren en ambas listas.
- El vector intersección se mostrará por pantalla siempre que existan elementos comunes.

NOTA: Se comprobará que el usuario no introduzca elementos repetidos dentro del mismo vector.


#### Ejemplo de ejecución:
<pre>
Introduce el numero de elementos de la lista 1 (>0):  3
Introduce los números:
Dato 1: 2
Dato 2: 4
Dato 3: 1

Introduce el numero de elementos de la lista 2 (>0):  5
Introduce los números:
Dato 1: 3
Dato 2: 3
Error, elemento repetido. Dato 2: 2
Dato 3: 9
Dato 4: 0
Dato 5: 1

La intersección de las listas es: 2 1

-----------------------------------------------------------

Introduce el numero de elementos de la lista 1 (>0):  2
Introduce los números:
Dato 0: 3
Dato 1: -2

Introduce el numero de elementos de la lista 2 (>0):  3
Introduce los números:
Dato 0: 0
Dato 1: 4
Dato 2: 5

No hay elementos comunes
</pre>

```python
# Realiza tu solución aquí o en un archivo .py independiente

def leer_lista(numero_lista):
    while True:
        try:
            n = int(input(f"Introduce el numero de elementos de la lista {numero_lista} (>0): "))
            if n <= 0:
                raise ValueError("El número debe ser mayor que 0.")
            break
        except ValueError as e:
            print(e)

    lista = []
    print("Introduce los números:")
    for i in range(n):
        while True:
            try:
                dato = int(input(f"Dato {i + 1}: "))
                if dato in lista:
                    raise ValueError("Error, elemento repetido.")
                lista.append(dato)
                break
            except ValueError as e:
                print(e)

    return lista

def main():
    print("\n=== Lectura de la primera lista ===")
    lista1 = leer_lista(1)

    print("\n=== Lectura de la segunda lista ===")
    lista2 = leer_lista(2)

    interseccion = list(set(lista1) & set(lista2))

    if interseccion:
        print(f"\nLa intersección de las listas es: {' '.join(map(str, interseccion))}")
    else:
        print("\nNo hay elementos comunes.")

if __name__ == "__main__":
    main()
```

Escribir un programa que cree **un conjunto** simulando una lista de la compra. 

El programa debe preguntar el artículo y añadirlo al conjunto, mostrando un mensaje de aviso en caso de que el artículo ya existiese.



#### Ejemplo de ejecución:

<pre>
Introduce un artículo: pan
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: agua
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: leche
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: café
¿Quieres añadir artículos a la lista (Si/No)? No
Artículos añadidos:
leche
agua
pan
café
Número de artículos:  4
</pre>

```python
# Realiza tu solución aquí o en un archivo .py independiente

def main():
    lista_compra = set()  

    while True:
        articulo = input("Introduce un artículo: ").strip()
        if articulo in lista_compra:
            print(f"El artículo '{articulo}' ya está en la lista.")
        else:
            lista_compra.add(articulo)
            print(f"Artículo '{articulo}' añadido a la lista.")

        continuar = input("¿Quieres añadir artículos a la lista (Si/No)? ").strip().lower()
        if continuar != 'si':
            break

    print("\nArtículos añadidos:")
    for articulo in lista_compra:
        print(articulo)

    print(f"Número de artículos: {len(lista_compra)}")

if __name__ == "__main__":
    main()
```

Implementar un programa en Python que:
- Lea una frase (f) por teclado.
- Lea un caracter por teclado (c) 
- Cambie todas las vocales de f por c Muestre f

#### Ejemplo de ejecución:

Dime una frase: Programar en python es muy divertido<br>
Dime un caracter: o<br>
La nueva frase es: Progromor on python os moy dovortodo<br>

================================================================

Dime una frase: Estamos trabajando con listas<br>
Dime un caracter: ?<br>
La nueva frase es: ?st?m?s tr?b?j?nd? c?n l?st?s<br>

```python
# Realiza tu solución aquí o en un archivo .py independiente

def main():
    frase = input("Dime una frase: ").strip()
    caracter = input("Dime un caracter: ").strip()

    while len(caracter) != 1:
        print("Por favor, introduce un único caracter.")
        caracter = input("Dime un caracter: ").strip()

    vocales = "aeiouAEIOU"
    nueva_frase = "".join([caracter if letra in vocales else letra for letra in frase])

    print(f"La nueva frase es: {nueva_frase}")

if __name__ == "__main__":
    main()
```

<h2 style="color: #007BC4;">Ejercicio 4</h2>
<hr style="border: 0.5px solid #007BC4;">

Realizar un programa en Python que permita gestionar las subscripciones de un usuario a diversas plataformas de entretenimiento, utilizando para ello **un diccionario**.

- El ejercicio parte de un diccionario que contiene el nombre y precio mensual de las subscripciones, formado por la clave (nombre de la plataforma) y el valor (precio mensual):


<pre>
        plataformas = {
        "Netflix": 11.99,
        "HBO": 8.99,       
        "Disney+": 7,
        "DAZN": 9.99,
        "Spotify": 9.99,
        "Apple Music": 9.99,
        "Amazon Prime": 3.99,
        }
</pre>


A continuación, deberá mostrar repetidamente un menú con las siguientes 6 opciones: 

1. Añadir plataforma (permite añadir una nueva plataforma al diccionario)
2. Eliminar plataforma (elimina una plataforma del diccionario)
3. Mostrar plataforma (muestra el precio de una plataforma a partir de su nombre)
4. Listar plataformas (muestra la lista de plataformas y sus precios)
5. Subscribirse n meses a una plataforma (muestra el precio total de la subscripción) 
6. Terminar (sale del programa)



#### Ejemplo de ejecución:

<img src="img/imagen1.png">

```python
# Realiza tu solución aquí o en un archivo .py independiente

def mostrar_menu():
    print("\n--- Menú de Opciones ---")
    print("1. Añadir plataforma")
    print("2. Eliminar plataforma")
    print("3. Mostrar plataforma")
    print("4. Listar plataformas")
    print("5. Subscribirse n meses a una plataforma")
    print("6. Terminar")

def anadir_plataforma(plataformas):
    nombre = input("Introduce el nombre de la plataforma: ").strip()
    if nombre in plataformas:
        print("La plataforma ya está registrada.")
    else:
        try:
            precio = float(input("Introduce el precio mensual de la plataforma: ").strip())
            if precio <= 0:
                raise ValueError("El precio debe ser mayor que 0.")
            plataformas[nombre] = precio
            print(f"Plataforma '{nombre}' añadida con éxito.")
        except ValueError:
            print("Error: el precio debe ser un número válido y mayor que 0.")

def eliminar_plataforma(plataformas):
    nombre = input("Introduce el nombre de la plataforma a eliminar: ").strip()
    if nombre in plataformas:
        del plataformas[nombre]
        print(f"Plataforma '{nombre}' eliminada con éxito.")
    else:
        print("La plataforma no está registrada.")

def mostrar_plataforma(plataformas):
    nombre = input("Introduce el nombre de la plataforma: ").strip()
    if nombre in plataformas:
        print(f"El precio mensual de '{nombre}' es {plataformas[nombre]:.2f} €.")
    else:
        print("La plataforma no está registrada.")

def listar_plataformas(plataformas):
    if plataformas:
        print("\nLista de plataformas y sus precios:")
        for nombre, precio in plataformas.items():
            print(f"- {nombre}: {precio:.2f} €")
    else:
        print("No hay plataformas registradas.")

def subscribirse_n_meses(plataformas):
    nombre = input("Introduce el nombre de la plataforma: ").strip()
    if nombre in plataformas:
        try:
            meses = int(input("Introduce el número de meses: ").strip())
            if meses <= 0:
                raise ValueError("El número de meses debe ser mayor que 0.")
            total = plataformas[nombre] * meses
            print(f"El precio total por {meses} meses de subscripción a '{nombre}' es {total:.2f} €.")
        except ValueError:
            print("Error: el número de meses debe ser un entero positivo.")
    else:
        print("La plataforma no está registrada.")

def main():
    plataformas = {
        "Netflix": 11.99,
        "HBO": 8.99,
        "Disney+": 7,
        "DAZN": 9.99,
        "Spotify": 9.99,
        "Apple Music": 9.99,
        "Amazon Prime": 3.99,
    }

    while True:
        mostrar_menu()
        opcion = input("Selecciona una opción: ").strip()

        if opcion == "1":
            anadir_plataforma(plataformas)
        elif opcion == "2":
            eliminar_plataforma(plataformas)
        elif opcion == "3":
            mostrar_plataforma(plataformas)
        elif opcion == "4":
            listar_plataformas(plataformas)
        elif opcion == "5":
            subscribirse_n_meses(plataformas)
        elif opcion == "6":
            print("Saliendo del programa. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, selecciona una opción del menú.")

if __name__ == "__main__":
    main()




```
